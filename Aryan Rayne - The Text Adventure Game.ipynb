```python
"""
author: Arundhishaan

"""
"""
    DocString:
    
    I) Introduction: Welcome to Aryan Rayne - The Text Adventure Game. This
    game is based on a fiction story called Aryan Rayne. Aryan Rayne is chosen
    to defeat Omega Pearl, the ultimate villain.
    
    II) The game has three stages:
        Stage 1: Kill Jack the Reaper - By finding the magic word!
        Stage 2: Save Miss Riyana Rouzelle from the Ogra Clan - SHAPE SHIFTERS.
        Stage 3: Defeat Omega Pearl - Selecting the best weapon!

""" 
##############################################################################

from sys import exit

def fail():
    """The player failed the game"""
    print("You failed! Aryan Rayne is dead. Try harder again!")
    
    while True:
        answer = input("Do you want to play again? yes/no : ").casefold()
        if answer == 'yes':
            return stage_1()
        elif answer == 'no':
            print('Sorry to hear that, bye!')
            exit()
        print('I didn\'t understand you, try again')

##############################################################################        
        
def stage_1():
    """Stage 1: Kill Jack the Reaper - By finding the magic word!"""
    
    # Game Starts 
    print(f"""
    {'*'*40}\n
    Wow! My savior has arrived. I am Aryan Rayne. My Quest is to 
    defeat Omega Pearl. I can use your help!\n
    {'*'*40} 
    """)
    
    pl_name = input('Player name:  ')
    
    print(f"""
    {'*'*40}\n      
    Hello {pl_name}!
    
    Ohh! This is Jack the Reaper of Souls! One of Omega Pearl's henchman.
        
    Jack the Reaper: Aryan Rayne, the nemesis of my master. You will never 
    get past me.\n
    {'*'*40}
    """)
    
    input("<Press any key to continue>")
    
    # Stage 1, to find the missing word
    
    print(f"""
    {'>'*40}\n 
    The only way to defeat Jack is by finding the magic word from 
    my magic book!\n
    {'<'*40}
    """)
    
    input("<Press any key to continue>")
    
    print(f"""
    Letters are missing!! Please find them before impatient Jack kills me!
          
    {'~'*80}\n
                ╔╦╗┌─┐┌─┐┬┌─┐  ╔╗ ┌─┐┌─┐┬┌─
                ║║║├─┤│ ┬││    ╠╩╗│ ││ │├┴┐
                ╩ ╩┴ ┴└─┘┴└─┘  ╚═╝└─┘└─┘┴ ┴ \n
    {'~'*80}\n
    
    ___  E  A  ___  ___ T  S
    (1)        (2)  (3)    
          
    Squirrels like me! But Reapers don't, 
            Shoot the P, and Jack goes Nuts!!
          
    """)
    
    # Timed conversation while player thinks to enter input
    
    import time
    from threading import Thread
  
    # Get first missing letter 
 
    answer_1 = None
    
    def check_1():
        """Six second timer to show message and prompt user to enter value faster"""
        time.sleep(6)
        if answer_1 != None:
            return 
        print (f"""
    Hurry up, my friend. Jack knows everything what you are thinking!
    He is wierd and crazy! He might do something! 
        """)
      
    Thread(target = check_1).start()

    answer_1 = input("Enter the first letter into the magic book:  ")
    
    answer_1 = answer_1.upper()   

    # Giving chances to guess right
    
    chance_1 = 1

    while chance_1>=0:
        
        if answer_1 == 'P':
            print(f"""
        {'='*40}\n
          
            CORRECT!!!  
        
        P  E  A  ___  ___ T  S
       (1)       (2)  (3)    
          
    Squirrels like me! But Reapers don't, 
            Shoot the P, and Jack goes Nuts!!\n
        {'='*40}\n  
            """)
            break
            
        elif chance_1 == int(0):
            chance_1-=1
            fail()
    
        elif answer_1 != 'P':
            answer_1 = input("\n That's Wrong! Jack is coming, try again: ")
            answer_1 = answer_1.upper()
            chance_1-=1
            
        else:
            print("Something went wrong!")
  
    # Get second missing letter 

    answer_2 = None
    
    def check_2():
        """Three second timer to show message and prompt user to enter value faster"""
        time.sleep(3)
        if answer_2 != None:
            return 
        print (f"""
    Almost there, come on you can do this. I know you can! 
        """)
      
    Thread(target = check_2).start()

    answer_2 = input("Now, enter the second letter into the magic book: ")
    
    answer_2 = answer_2.upper()
    
    if answer_2 != 'N':
        print("That is incorrect!")
        fail()
    
    elif answer_2 == 'N':
        print(f"""
    {'='*40}\n   
          You are AWESOME!!!!!\n
        
        P  E  A   N  ___  T  S
       (1)       (2) (3)    
          
    Squirrels like me! But Reapers don't, 
            Shoot the P, and Jack goes Nuts!!\n
    {'='*40}\n  
        """)
     
    # Get third missing letter

    answer_3 = None
    
    def check_3():
        """Five second timer to show message and prompt user to enter value faster """
        time.sleep(5)
        if answer_3 != None:
            return 
        print (f"""
    Faster my friend! Jack is on to me...........
        """)
      
    Thread(target = check_3).start()

    answer_3 = input("Last one, enter the third letter into the magic book: ")

    answer_3 = answer_3.upper()

    if answer_3 == 'U':
        print(f"""
    {'^'*40}\n 
          Yes, great going!!!
          
    Squirrels like me! But Reapers don't, 
            Shoot the P, and Jack goes Nuts!!
          
         >>>>> P E A N U T S <<<<<
          
    
    {'*'*15} YES!! Jack the Reaper is dead. Lets keep going! {'*'*15}
          
    {'^'*40}\n
    
        """)
        input("<Press any key to continue>")
        stage_2()
    
    else:
        fail()
        
##############################################################################

def stage_2():
    """Stage 2: Save Miss Riyana Rouzelle from the Ogra Clan - SHAPE SHIFTERS"""
    
    # Encounter with Miss Riyana Rouzelle
    
    print(f"""
    {'*'*40}\n
    Look who's here!! It is Miss Riyana Rouzelle, princess of our Kingdom. 
    She is captured by the Ogra Clan.
    
    Be careful, they are SHAPE SHIFTERS! \n
    {'*'*40}\n
    """)
    
    input("<Press any key to continue>")
    
    print(f"""
    {'>'*40}\n
    I need to answer Miss Riyana's question to prove myself among the 
    shape shifters and save her.\n
    {'<'*40}\n  
          
    Should I save Miss Riyana by answering or go defeat Omega Pearl 
    to save the world?
    
    A. Save Miss Riyana
          
    B. Go save the world
          
    C. Keep Thinking..... \n
          
    """)    
    
    reply_1 = input("What should I do now?  >>  ")
    
    reply_1 = reply_1.upper()
    
    reply_lst = ['A','SAVE MISS','SAVE RIYANA','RIYANA', 'SAVE']
    no_reply_lst = ['B', 'C','WORLD','THINKING','THINK']
    
    if reply_1 in reply_lst:
        print(f"""
        {'^'*40}\n
              Alright, I've got this! Leave it to me.
          
        ==>   Miss Riyana: What is the first couplet that you wrote for me?
            
              [*couplet: 2 line poem]
          
        {'^'*40}\n  
        """)
        
        input("<Press any key to hear the answer>")
        
        print(f"""
        Aryan Rayne:  || You always say me never,
                              The day I never saw You.  ||
          
        {'*'*15} Yes, Good Decision! We saved Miss Riyana Rouzelle. {'*'*15} \n
        """)
        
        input("<Press any key to continue>")
        stage_3()
        
    elif reply_1 in no_reply_lst:
        print(f"""
        Ohh... No! We are surrounded by the shape shifters! \n
        """)
        fail()
        
    else:
        print(f"""
        Something went wrong!""")
        fail()

##############################################################################

def stage_3():
    """Stage 3: Defeat Omega Pearl! - Selecting the best weapon!"""
    
    print(f"""
    {'*'*40}\n
    Well, here we are! It is now or never. 
          
    Get ready for battle!! \n
     {'*'*40}\n
          
    <<< BOSS FIGHT >>> """)
    
    input ("<Enter any key to continue>")
    
    # Selecting weopons to defeat Omega Pearl
    
    import random

    health_omega = 100
    
    equip_1 = [['SWORD', 30], ['DAGGER', 20], ['SPEAR', 50], ['FLUTE', 70]]
    
    print(f"""
    These are your weapons, ||Weapon, Strike Points|| : {equip_1} 

    Health of Omega Pearl  : {health_omega} 

    You have 3 shots in a row. Use it wisely. 
    
    This is it, my friend. Lets make it happen!!
    """)
    
    input("<Press any key to continue>")
    
    shots = []
    chances = 3 

    while chances>0 and health_omega>0:    
    
        tool_1 = input("Which weopon would you like to use? Sword, Dagger, Spear or Flute?  >>     ")
        
        tool_1 = tool_1.upper()
        chances-=1
        
        for tool, points in equip_1:
            if tool_1 == tool:
                health_omega -= points
                shots.append(tool)
                print(f"""
    {'-'*40}\n
    Shots              : {shots} 
    Shots remaining    : {chances}
    Omega Pearl Health : {health_omega} pts
    Health lost        : {points} 
    {'-'*40}
                """)
                continue
            
    if health_omega <= 0:
        print(f"""
                  
    {'*'*15} Omega Pearl is dead! Ultimate Victory! {'*'*15} \n               

          +-+-+   -+-+-+-+-    +-+    -+-+-+-+-+-+-
          |W|E|   |H|A|V|E|    |A|    |W|I|N|N|E|R|
          +-+-+   -+-+-+-+-    +-+    -+-+-+-+-+-+-      

    {'>'*20} Thank you for playing Aryan Rayne - The Text Adventure Game {'<'*20} 
    
        """)
        input("<Press any key>")
        exit()

    elif health_omega > 0:
        print(f"""
    You have {chances} shots remaining.
    """)
        fail()
    
    else:
        print("You have {chances} shots remaining!")
        fail()

stage_1()
```
